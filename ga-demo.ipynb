The ratio of area to the square of the perimeter, $\text{Area} / \text{Perimeter}^2$, is a measure of the compactness or efficiency of a shape. Here’s a detailed explanation:

### Area

The area of a polygon is the measure of the space enclosed by the polygon. For a given set of vertices, the area can be calculated using the Shoelace formula (Gauss's area formula).

### Perimeter

The perimeter of a polygon is the total length of its sides. For a given set of vertices, the perimeter can be calculated by summing the Euclidean distances between consecutive vertices.

### $\text{Area} / \text{Perimeter}^2$

This ratio is used to assess how effectively a polygon encloses its area relative to its perimeter. 

- **Higher Values**: A higher value indicates that the shape is more compact and efficient at enclosing area relative to its perimeter. For example, a circle has a high ratio because it encloses the maximum area for a given perimeter length.
  
- **Lower Values**: A lower value indicates that the shape is less efficient at enclosing area relative to its perimeter. Shapes with many sides or that are highly irregular will typically have lower ratios.

### Why Use This Ratio?

- **Compactness**: It’s useful for understanding the compactness of a shape. More compact shapes are often more desirable in various applications such as materials science, biology, and urban planning.
  
- **Efficiency**: In some optimization problems, this ratio can help find shapes that maximize area while minimizing perimeter, leading to efficient use of resources.

### Example Calculation

For a given polygon with vertices, the steps to calculate $\text{Area} / \text{Perimeter}^2$ are:

1. **Calculate the Area**: Use the Shoelace formula or another appropriate method.
2. **Calculate the Perimeter**: Sum the Euclidean distances between consecutive vertices.
3. **Compute the Ratio**: Divide the area by the square of the perimeter.

### Implementation in the Code

In the provided code, the fitness function computes this ratio to evaluate each polygon:

```python
def fitness(polygon):
    area, perimeter = calculate_area_and_perimeter(polygon)
    if perimeter == 0:
        return 0
    return area / (perimeter ** 2)
```

- **Area Calculation**: `calculate_area_and_perimeter` computes the area of the polygon.
- **Perimeter Calculation**: The same function computes the perimeter of the polygon.
- **Fitness Value**: The function returns the ratio $\text{Area} / \text{Perimeter}^2$.

This ratio is then used to guide the genetic algorithm in selecting, crossing, and mutating shapes to find the most compact and efficient polygon.

````python
import numpy as np
import matplotlib.pyplot as plt

from point import Point
from polygon import Polygon
from ga import GeneticAlgorithm
````

## 1. Randomly sample points

````python
n_points = 10
points = [Point(np.random.rand(), np.random.rand()) for _ in range(n_points)]

x = [point.x for point in points]
y = [point.y for point in points]
plt.plot(x, y, 'o')
plt.title('Randomly sampled points')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal');
````

## 2. Sample a polygon and determine the fitness

From these 5 points we can randomly sample a polygon and determine $\text{Area} / \text{Perimeter}^2$.

````python
import random

random.shuffle(points)
polygon = Polygon(points)
polygon.plot()
print(f"Fitness score: {polygon.fitness():.4f}")
````

## 3. Initial population

````python
ga = GeneticAlgorithm(points, population_size=25, num_generations=50, num_parents=4)
ga.plot_population()
````

## 4. Optimisation

````python
best_polygon = ga.evolve()
best_polygon.plot()
````
